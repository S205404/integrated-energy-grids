```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True

def f(x):
   return (x-5)**2 

x = np.linspace(0, 10, 100)

plt.plot(x, f(x), color='blue')
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,10])
plt.ylim([0,25])
plt.savefig('fig1.png')
```

```python
plt.rcParams["figure.figsize"] = [3, 3]
plt.rcParams["figure.autolayout"] = True

def f(x):
   return (x-5)**2
x = np.linspace(0, 10, 100)
plt.plot(x, f(x), color='blue')


colors=['gold', 'darkorange', 'red']
for i,k in enumerate([2, 0.5, 0.1]):
    def f(x):
       return (x-5)**2 -k*np.log(x-6)
    plt.plot(x, f(x), color=colors[i], label=k)
    
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.xlim([0,10])
plt.ylim([0,25])
plt.legend(title='k')
plt.savefig('fig4.png')
```

```python
plt.rcParams["figure.figsize"] = [4, 3]
plt.rcParams["figure.autolayout"] = True
r=0.0001
def f(x):
   return r*x**2 

x = np.linspace(-600, 600, 100)

plt.plot(x, f(x), color='blue')
plt.fill_between(x, f(x), 8.5, where=(f(x) <8.5), color='blue', alpha=0.6)

plt.grid()
plt.xlabel('$p_l$ (MW)')
plt.ylabel('$\Psi_l$ (MW)')
plt.ylim([-2,15])
plt.xlim([-600,600])
plt.savefig('fig10.png')
```

```python
plt.rcParams["figure.figsize"] = [4, 3]
plt.rcParams["figure.autolayout"] = True
ro = 983 #kg/m3
f_d = 0.025

def f(m,d):
   return f_d*1*m**2/(2*ro*d**5*np.pi**2)
    
#m=np.array([0.001, 0.1, 1, 10, 100, 1000, 10000])
m = np.linspace(0, 10000, 20000)
for d in [2, 5, 10, 20, 50, 100]: #, 0.1, 0.05]:
        plt.plot(m, f(m,d/100), label=d)
plt.grid()
plt.xlabel('Mass flow (kg/s)')
plt.ylabel('Pressure drop (Pa/m)')
plt.legend(title="diameter (cm) ",)
plt.xlim([0.01,10000])
plt.ylim([0,300])
plt.xscale('log')
plt.savefig('fig14.png')
```

```python
plt.rcParams["figure.figsize"] = [4, 3]
plt.rcParams["figure.autolayout"] = True

def f(d,m):
    return 100/(1+d)**(m-2020)

m = np.linspace(2020.1, 2100, 100)
for d in [0, 0.02, 0.05, 0.1]:
        plt.plot(m, f(d,m), label=str(round(d*100))+'%')
plt.grid()

plt.ylabel('Present value of 100 EUR')
plt.legend(title="discount rate",)
plt.xlim([2020,2100])
plt.ylim([0,105])
plt.savefig('fig18.png')
```
